#Hashtag Recommendation System for Social Media Text

This project generates relevant hashtags from input text using basic Natural Language Processing techniques

```python
import re

texts = [
    "I love coding in Python",
    "Machine learning is amazing",
    "Data science is the future",
    "Artificial intelligence and deep learning",
    "Python programming is fun"
]

def generate_hashtags(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    words = text.split()

    stopwords = ["is", "the", "in", "and", "to", "a", "of","i"]

    keywords = []
    for word in words:
        if word not in stopwords:
            keywords.append(word)

    hashtags = []
    for word in keywords:
        hashtags.append("#" + word)

    return hashtags


for text in texts:
    print("Text:", text)
    print("Hashtags:", generate_hashtags(text))
    print("------")
```

## How it Works

- Text is converted to lowercase
- Special characters are removed
- Words are filtered by removing common stopwords
- Remaining keywords are converted into hashtags

```python
# User input option
user_text = input("Enter your text: ")
print("Generated Hashtags:", generate_hashtags(user_text))
```
