#Sentiment Analysis on Social Media Text
This project analyzes the sentimenmt of text data (positive ,negative ,neutral) using Natural Language Prcessing (NLP).

It uses the TextBlob library to determine Sentiment Polarity and classify text accordingly .

```python
!pip install textblob
```

```python
from textblob import TextBlob
# Sample social media texts
texts=[
    "I  love this product ",
    "This is the worst experience ever",
    "it's okay ,not bad ",
    "Absolutely amazing sevice",
    "I hate it "
]

#Function to classify sentiment
def get_sentiment(text):
  analysis=TextBlob(text)

  if analysis.sentiment .polarity > 0 :
    return "Positive"
  elif analysis .sentiment.polarity<0:
    return "Negative"
  else:
    return "Neutral"
#process all texts
for text in texts :
  sentiment= get_sentiment(text)
  print(f"Text: {text}")
  print(f"Sentiment: {sentiment}")
  print("-----")
```

##How it works
-The input text is processed using the TextBlob library
-Sentiment polarity is calculated
-Based on polarity :
  - Positive (>0)
  - Negative (<0)
  - Neutral  (=0)

-The result is printed for each input text
